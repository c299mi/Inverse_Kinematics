```python
import numpy as np
import matplotlib.pyplot as plt

# 各リンクの長さ
L1 = 0.522
L2 = 0.83
L3 = 0.935
L4 = 0.65
L5 = 0.6
L6 = 1.0

# 関節角度 (ラジアン)
theta1 = np.radians(90)  # L1は常に90度
theta2 = np.radians(-45)  # 例：-45度（相対角度）
theta3 = np.radians(120)  # 例：120度（相対角度）
theta4 = np.radians(30)   # 例：30度（相対角度）
theta5 = np.radians(270)  # θ5は常に270度固定
theta6 = np.radians(90)   # θ6は常に90度固定

# 順運動学の計算
# リンク1の終端座標
x1 = -L1 * np.cos(theta1)  # 左右反転のため符号を変更
y1 = L1 * np.sin(theta1)

# リンク2の終端座標
x2 = x1 - L2 * np.cos(theta1 + theta2)  # 左右反転のため符号を変更
y2 = y1 + L2 * np.sin(theta1 + theta2)

# リンク3の終端座標
x3 = x2 - L3 * np.cos(theta1 + theta2 + theta3)  # 左右反転のため符号を変更
y3 = y2 + L3 * np.sin(theta1 + theta2 + theta3)

# リンク4の終端座標
x4 = x3 - L4 * np.cos(theta1 + theta2 + theta3 + theta4)  # 左右反転のため符号を変更
y4 = y3 + L4 * np.sin(theta1 + theta2 + theta3 + theta4)

# リンク5の終端座標
x5 = x4 - L5 * np.cos(theta1 + theta2 + theta3 + theta4 + theta5)  # 左右反転のため符号を変更
y5 = y4 + L5 * np.sin(theta1 + theta2 + theta3 + theta4 + theta5)

# リンク6の終端座標（エンドエフェクタの座標）
x6 = x5 - L6 * np.cos(theta1 + theta2 + theta3 + theta4 + theta5 + theta6)  # 左右反転のため符号を変更
y6 = y5 + L6 * np.sin(theta1 + theta2 + theta3 + theta4 + theta5 + theta6)

# 図のプロット
plt.figure(figsize=(8, 8))
plt.plot([0, x1, x2, x3, x4, x5, x6], [0, y1, y2, y3, y4, y5, y6], 'ro-')
plt.scatter([0, x1, x2, x3, x4, x5, x6], [0, y1, y2, y3, y4, y5, y6], color='blue')
plt.text(0, 0, 'O', fontsize=12, ha='right')
plt.text(x1, y1, 'A', fontsize=12, ha='right')
plt.text(x2, y2, 'B', fontsize=12, ha='right')
plt.text(x3, y3, 'C', fontsize=12, ha='right')
plt.text(x4, y4, 'D', fontsize=12, ha='right')
plt.text(x5, y5, 'E', fontsize=12, ha='right')
plt.text(x6, y6, 'F', fontsize=12, ha='right')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('kinematics')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# 各リンクの終端座標を出力
print(f"リンク1の終端座標: ({x1:.3f}, {y1:.3f})")
print(f"リンク2の終端座標: ({x2:.3f}, {y2:.3f})")
print(f"リンク3の終端座標: ({x3:.3f}, {y3:.3f})")
print(f"リンク4の終端座標: ({x4:.3f}, {y4:.3f})")
print(f"リンク5の終端座標: ({x5:.3f}, {y5:.3f})")
print(f"エンドエフェクタの座標: ({x6:.3f}, {y6:.3f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_robot(L1, L2, L3, L4, L5, L6, theta2, theta3, theta4):
    # 各リンクの長さ
    L1 = 0.522
    L2 = 0.83
    L3 = 0.935
    L4 = 0.65
    L5 = 0.6

    # 関節角度 (ラジアン)
    theta1 = np.radians(90)  # L1は常に90度
    theta2 = np.radians(theta2)  # 例：-45度（相対角度）
    theta3 = np.radians(theta3)  # 例：120度（相対角度）
    theta4 = np.radians(theta4)  # 例：30度（相対角度）
    theta5 = np.radians(270)  # θ5は常に90度固定
    theta6 = np.radians(90)  # 任意の角度

    # 順運動学の計算
    # リンク1の終端座標
    x1 = -L1 * np.cos(theta1)  # 左右反転のため符号を変更
    y1 = L1 * np.sin(theta1)

    # リンク2の終端座標
    x2 = x1 - L2 * np.cos(theta1 + theta2)  # 左右反転のため符号を変更
    y2 = y1 + L2 * np.sin(theta1 + theta2)

    # リンク3の終端座標
    x3 = x2 - L3 * np.cos(theta1 + theta2 + theta3)  # 左右反転のため符号を変更
    y3 = y2 + L3 * np.sin(theta1 + theta2 + theta3)

    # リンク4の終端座標
    x4 = x3 - L4 * np.cos(theta1 + theta2 + theta3 + theta4)  # 左右反転のため符号を変更
    y4 = y3 + L4 * np.sin(theta1 + theta2 + theta3 + theta4)

    # リンク5の終端座標
    x5 = x4 - L5 * np.cos(theta1 + theta2 + theta3 + theta4 + theta5)  # 左右反転のため符号を変更
    y5 = y4 + L5 * np.sin(theta1 + theta2 + theta3 + theta4 + theta5)

    # リンク6の終端座標（エンドエフェクタの座標）
    x6 = x5 - L6 * np.cos(theta1 + theta2 + theta3 + theta4 + theta5 + theta6)  # 左右反転のため符号を変更
    y6 = y5 + L6 * np.sin(theta1 + theta2 + theta3 + theta4 + theta5 + theta6)

    # 図のプロット
    plt.figure(figsize=(8, 8))
    plt.plot([0, x1, x2, x3, x4, x5, x6], [0, y1, y2, y3, y4, y5, y6], 'ro-')
    plt.scatter([0, x1, x2, x3, x4, x5, x6], [0, y1, y2, y3, y4, y5, y6], color='blue')
    plt.text(0, 0, 'O', fontsize=12, ha='right')
    plt.text(x1, y1, 'A', fontsize=12, ha='right')
    plt.text(x2, y2, 'B', fontsize=12, ha='right')
    plt.text(x3, y3, 'C', fontsize=12, ha='right')
    plt.text(x4, y4, 'D', fontsize=12, ha='right')
    plt.text(x5, y5, 'E', fontsize=12, ha='right')
    plt.text(x6, y6, 'F', fontsize=12, ha='right')
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title('6link（L1が垂直）')
    plt.xlabel('X座標')
    plt.ylabel('Y座標')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

    # 各リンクの終端座標を出力
    print(f"リンク1の終端座標: ({x1:.3f}, {y1:.3f})")
    print(f"リンク2の終端座標: ({x2:.3f}, {y2:.3f})")
    print(f"リンク3の終端座標: ({x3:.3f}, {y3:.3f})")
    print(f"リンク4の終端座標: ({x4:.3f}, {y4:.3f})")
    print(f"リンク5の終端座標: ({x5:.3f}, {y5:.3f})")
    print(f"エンドエフェクタの座標: ({x6:.3f}, {y6:.3f})")

# 関数の呼び出し（例）
L6 = 1.0  # 任意の長さ
theta2 = -45  # 任意の角度（度）
theta3 = 120  # 任意の角度（度）
theta4 = 30   # 任意の角度（度）
theta6 = 90   # 任意の角度（度）

calculate_and_plot_robot(0.522, 0.83, 0.935, 0.65, 0.6, L6, theta2, theta3, theta4)
```
